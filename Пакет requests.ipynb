1) переходим по первой ссылке (из первого input) и собираем все ссылки с полученной страницы в первый массив; 
2) переходим по каждой из собранных ссылок и собираем ссылки, на каждой странице, во второй массив;
3) проверяем вхождение второй введенной ссылки (из второго input) во второй массив и печатаем Yes, если она там есть.

Поэтому, первым шагом импортируем requests и re.
Вводим две переменные через ``input().replace('stepic.org', 'stepik.org')``, назовем их A и B.

Создаем пустой список, допустим, ``all_links``.
Считываем A в переменную ``site`` через ``requests.get(A).text``.
Ищем все ссылки сайта A с помощью ``re.findall("href=[\"\'](.*?)[\"\']", site)`` и сохраняем результат в переменную ``links``.

Циклом проходимся по ``links``, где каждый элемент (link) списка считываем с помощь ``requests.get(link.replace('stepic.org', 'stepik.org')).text`` и сохраняем в переменную, в моем случае, названную ``html``.

``re.findall("href=[\"\'](.*?)[\"\']", html)`` в переменную ``cycle_sites``.
Чтобы не было проблем со ('stepic.org', 'stepik.org'): ``cycle_sites = list(map(lambda x: x.replace('stepic.org', 'stepik.org'), cycle_sites))``

Добавляем новые ссылки в наш раннее обозначенный список: ``all_links.extend(cycle_sites)``
Выходим из цикла и делаем проверку: если В в all_links выводим 'Yes', иначе 'No'.

```python
# put your python code here
import requests
import re
f=False
pattern = r'(?:href=[\"\'])(.*?)(?:[\"\'])'
x1, x2=[input().replace('stepic.org', 'stepik.org') for _ in range(2)]
t1=requests.get(x1)

z1=re.findall(pattern, t1.text)
z1=list(map(lambda i: i.replace('stepic.org', 'stepik.org'),z1))


z3=[]
for i in z1:
    try:
        z3.extend(list(map(lambda j: j.replace('stepic.org', 'stepik.org'),re.findall(pattern, requests.get(i).text))))
        if x2 in z3:
            f=True
            break
    except:
        pass

if x2 in z3:
    f=True
    
if f: print('Yes')
else: print('No')
```
