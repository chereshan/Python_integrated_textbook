```python
import string
from random import *
d1=[i for i in list(string.ascii_letters) if i not in 'lI1oO0']
d2=[i for i in list(string.digits) if i not in 'lI1oO0']
d=d1+d2
def generate_password(length):
    t=sample(d, length)
    while not all(any(map(lambda x: x in string.ascii_lowercase,t)), any(map(lambda x: x in string.ascii_uppercase,t)), any(map(lambda x: x in string.digits,t))):
        t=t=sample(d, length)
    return ''.join(t)

def generate_passwords(count, length):
    for i in range(count):
        print(generate_password(length=length))

n, m = int(input()), int(input())
generate_passwords(n, m)
```
